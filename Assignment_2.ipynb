**Q4.** Create a function heart

‚óè

‚óè

Print both returned values separately.

image.png

```python
def heart_rate_stats(max_bpm, avg_bpm, resting_bpm):
    # Calculate Heart Rate Reserve
    heart_rate_reserve = max_bpm - resting_bpm

    # Calculate Average Difference
    average_difference = avg_bpm - resting_bpm

    # Return both values
    return heart_rate_reserve, average_difference


# Example usage
max_bpm = 190
avg_bpm = 130
resting_bpm = 70

# Get the results
hr_reserve, avg_diff = heart_rate_stats(max_bpm, avg_bpm, resting_bpm)

# Print the results separately
print("Heart Rate Reserve:", hr_reserve)
print("Average Difference:", avg_diff)
```

**Q6.** Create two functions:

‚óè
add
_
calories(c1, c2) that returns sum of calories.

‚óè
weekly_
burn() that calls add
_
calories() 7 times for each day‚Äôs burn.
Print total burn.

image.png

```python
# Function to add calories burned
def add_calories(c1, c2):
    # Return sum of calories
    return c1 + c2

# Function to calculate weekly calories burned
def weekly_burn(week):
    total = 0
    for day in week:
        total += add_calories(day[0], day[1])
    return total

# Example data: calories burned in two sessions per day for 7 days
week_data = [
    [1, 2],   # Day 1
    [3, 4],   # Day 2
    [1, 2],   # Day 3
    [5, 6],   # Day 4
    [3, 4],   # Day 5
    [6, 7],   # Day 6
    [9, 2]    # Day 7
]

# Print total weekly calories burned
print("Total Weekly Burn:", weekly_burn(week_data))
```

**Q12.** Write a function daily_
burns(calories
_
list) that prints burn for each day and returns total.

image.png

```python
def daily_burns(calories_list):
    total = 0
    day_num = 0
    week = ['Monday', 'Tuesday', 'Wednesday', 'Thrusday', 'Friday', 'Saturday', 'Sunday']

    # Loop through each day's calories
    for calories in calories_list:
        print(f"{week[day_num]} burn: {calories} calories")
        total += calories
        day_num += 1

    # Return total burn at the end
    return total


# Example usage
week_calories = [200, 300, 250, 400, 350, 500, 450]
total_burn = daily_burns(week_calories)

print("\nTotal Weekly Burn:",  total_burn)
```

**Q17.** Write a recursive function sum
_
n(n) to return sum of first n natural numbers.
Call sum
_
n(100).

image.png

```python
def sum_n(n):
    # Base case
    if n == 1:
        return 1
    else:
        # Recursive case
        return n + sum_n(n - 1)

print("Sum of first 100 natural numbers:", sum_n(100))
```

**Q18.** Write a recursive function factorial(n) to return n!
Call factorial(6).
image.png

```python
def factorial(n):
  # Base case
  if n == 1:
    return 1
  else:
    # Recursive case
    return n * factorial(n-1)

print(f"Factorial of {6} is {factorial(6)}")
```

```python
def top_5_burners(filename):
    # Read the dataset
    data = pd.read_csv(filename)

    # Sort data by 'Calories' column in descending order
    top_5 = data.sort_values(by='Calories', ascending=False).head(5)

    # Return top 5 rows
    return top_5

# Example usage
top_burners = top_5_burners('Final_data.csv')
print(top_burners)
```

**Q23.** Write a function to visualize Calories
Burned vs Session
Duration for each Workout
_
_
_
Matplotlib.
image.png

```python
def plot_calories_vs_duration(df):
    # Create scatter plot for each Workout_Type
    types = df['Workout_Type'].unique()
    for t in types:
        subset = df[df['Workout_Type'] == t].copy() # Use .copy() to avoid SettingWithCopyWarning
        # Convert hours to minutes for plotting
        subset['Session_Duration_minutes'] = subset['Session_Duration (hours)'] * 60
        plt.scatter(subset['Session_Duration_minutes'], subset['Calories_Burned'], label=t)

    plt.title("Calories Burned vs Session Duration")
    plt.xlabel("Session Duration (min)")
    plt.ylabel("Calories Burned")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example call
plot_calories_vs_duration(df)
```

**Q24.** Write a function to compute correlation between any two numeric columns (column names taken
from user input).

image.png

```python
def column_correlation(df):
    # Ask user to input column names
    col1 = input("Enter first column name: ")
    col2 = input("Enter second column name: ")

    # Check if both columns exist in the DataFrame
    if col1 in df.columns and col2 in df.columns:
        # Check if both columns are numeric
        if pd.api.types.is_numeric_dtype(df[col1]) and pd.api.types.is_numeric_dtype(df[col2]):
            corr = df[col1].corr(df[col2])
            print(f"Correlation between '{col1}' and '{col2}': {round(corr, 2)}")
        else:
            print("One or both columns are not numeric.")
    else:
        print("Invalid column names. Please check spelling and try again.")


column_correlation(df)
```

**Q25.** Write a function that creates and saves any Matplotlib plot as a PNG file automatically.

```python
def save_plot_as_png(x, y, filename='workout_plot.png'):
    plt.plot(x, y, marker='o')
    plt.title("Workout Summary")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.savefig(filename)
    print(f"Plot saved as {filename}")


save_plot_as_png([1, 2, 3, 4], [100, 200, 250, 300])
```
